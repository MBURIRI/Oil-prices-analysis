```python
data_1 = {
    'school': ['GP', 'GP', 'MS', 'MS'],
    'address': ['U', 'R', 'U', 'R'],
    'count': [631,141,144,128]
}

fam_size = pd.DataFrame()
sns.set(style="whitegrid")
palette = {'R': 'black', 'U': 'red'}
plt.figure(figsize=(8, 5))
bar_plot = sns.barplot(data=data_1, x='school', y='count', hue='address',palette=palette)
plt.title('Address of students for each school')
plt.xlabel('School')
plt.ylabel('Count')
plt.legend(title='address')
plt.show()
```

<h2 style="color: blue;">Mother's job</h2>

```python
df['Mjob'].value_counts()
```

```python
Mjob_data={'Mjob': ['other', 'services', 'teacher', 'at_home', 'health'],
    'count': [399, 239, 194, 130, 82]}
df_0 = pd.DataFrame(Mjob_data)


plt.figure(figsize=(5, 5))
plt.pie(df_0['count'], labels=df_0['Mjob'], autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))              
plt.title("Mother's Job Types Distribution", fontsize=16, fontweight='bold')
plt.axis('equal')
plt.tight_layout()
plt.legend()
plt.show()
```
