Decriptive Statistics


Descriptive statistics are a set of techniques used to summarize and describe the main features of a dataset. They provide a way to present data in a meaningful way through measures of central tendency, such as mean, median, and mode, which indicate the average or typical value; measures of variability, such as range, variance, and standard deviation, which show how spread out the data is; and measures of distribution, such as skewness and kurtosis, which reveal the shape and nature of the data's spread. By offering a clear and concise overview, descriptive statistics help to simplify complex data and reveal underlying patterns, making it easier to understand and interpret the information.

Measures of central tendancy


Measures of central tendency are statistical metrics used to identify the center point or typical value within a dataset. The primary measures include:

1. **Mean**: The arithmetic average of all values in a dataset, calculated by summing the values and dividing by the number of observations. It provides a general sense of the "center" but can be affected by extreme values or outliers.

2. **Median**: The middle value when a dataset is ordered from smallest to largest. If there is an even number of observations, it is the average of the two middle values. The median is less sensitive to outliers and skewed data, making it useful for understanding the center in skewed distributions.

3. **Mode**: The value or values that occur most frequently in a dataset. A dataset may have one mode (unimodal), more than one mode (bimodal or multimodal), or no mode if no value repeats. The mode is particularly useful for categorical data to identify the most common category.

These measures offer different perspectives on the dataâ€™s central location, helping to provide a comprehensive view of its typical values.

Skewness assesses asymmetry; positive skew means a longer right tail, while negative skew indicates a longer left tail. Both metrics help in understanding the shape and spread of data beyond basic mean and variance.

<h2 style="color: green;">Probability distribution</h2>

Probability distributions describe how values of a random variable are distributed. They show the likelihood of different outcomes, with discrete distributions (e.g., binomial) dealing with distinct values and continuous distributions (e.g., normal) handling a range of values. These distributions are fundamental in statistics for modeling and analyzing data, helping to understand the probability of various outcomes and their relative frequencies in a given dataset.

Probability distributions are crucial in statistical analysis and inferential statistics. They come in various forms:

1. **Discrete Distributions**: Include distributions like the Binomial (for counting successes in fixed trials) and the Poisson (for modeling rare events over a fixed interval).

2. **Continuous Distributions**: Include distributions like the Normal (bell-shaped curve, useful for many natural phenomena), Exponential (for time between events in a Poisson process), and Uniform (where all outcomes are equally likely).

3. **Probability Density Function (PDF)**: For continuous distributions, it describes the likelihood of different outcomes, and the area under the curve represents probabilities.

4. **Cumulative Distribution Function (CDF)**: Provides the probability that a random variable will take a value less than or equal to a specified value.

5. **Moments**: These include mean (first moment), variance (second central moment), skewness (third central moment), and kurtosis (fourth central moment), offering insights into the distribution's shape and spread.

Understanding these distributions helps in hypothesis testing, prediction, and data modeling, providing a framework for analyzing random processes and making informed decisions based on statistical evidence.

<h2 style="color: green;">Uniform distribution</h2>

```python

uniform_data = np.random.uniform(low=0, high=10, size=1000)

df = pd.DataFrame(uniform_data, columns=['value'])
sns.set(style='whitegrid')
ax = df.plot(kind='density', figsize=(8, 6), xlim=(-1, 11), lw=2, color='royalblue')
ax.set_xlabel('Value', fontsize=12)
ax.set_ylabel('Density', fontsize=12)
ax.set_title('Density Plot of Uniform Data', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show
```

<h2 style="color: green;">Normal distribution</h2>

```python
Normal_data = np.random.normal(5,2, size=10000)

df = pd.DataFrame(Normal_data, columns=['value'])
sns.set(style='whitegrid')
ax = df.plot(kind='density', figsize=(8, 6), xlim=(-1, 11), lw=2, color='royalblue')
ax.set_xlabel('Value', fontsize=12)
ax.set_ylabel('Density', fontsize=12)
ax.set_title('Density Plot of Normally distributed Data', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show
```

<h2 style="color: green;">Binomial distribution</h2>

```python
# Generate binomially distributed data
fair_coin_flips = np.random.binomial(n=10, p=0.5, size=10000)

# Cross-tabulation of the counts
cross_tab = pd.Series(fair_coin_flips).value_counts().sort_index()
print(cross_tab)

# Create a DataFrame for plotting
df = pd.DataFrame(fair_coin_flips, columns=['value'])

# Plot histogram
plt.figure(figsize=(8, 6))
df['value'].hist(range=(-0.5, 10.5), bins=11, color='royalblue', edgecolor='black', alpha=0.7)

# Add labels and title
plt.xlabel('Number of Heads', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.title('Histogram of Fair Coin Flips', fontsize=14)
# Add a grid
plt.grid(True, linestyle='--', alpha=0.7)

# Show plot
plt.show()
```
